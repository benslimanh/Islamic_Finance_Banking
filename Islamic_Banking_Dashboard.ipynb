```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. Manual Data Entry (Simulating a dataset to save time)
# Data represents 3 major Islamic banks over the last 3 years.
data = {
    'Year': [2022, 2023, 2024, 2022, 2023, 2024, 2022, 2023, 2024],
    'Bank': ['Al Rajhi (KSA)', 'Al Rajhi (KSA)', 'Al Rajhi (KSA)',
             'KFH (Kuwait)', 'KFH (Kuwait)', 'KFH (Kuwait)',
             'DIB (UAE)', 'DIB (UAE)', 'DIB (UAE)'],
    'Net_Profit': [3200, 4500, 5100, 900, 1400, 1900, 1100, 1350, 1600],  # In Millions USD
    'Deposits': [130, 155, 170, 50, 85, 95, 60, 65, 72],                 # In Billions USD
    'NPL_Ratio': [0.6, 0.5, 0.4, 1.2, 1.1, 1.0, 2.5, 2.1, 1.8]           # Non-Performing Loans % (Risk)
}

# Create a DataFrame
df = pd.DataFrame(data)

# 2. Dashboard Layout Configuration
# We create a 2x2 grid for our charts
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
plt.subplots_adjust(hspace=0.4) # Add space between rows
fig.suptitle('Islamic Banking Performance Dashboard (2022-2024)', fontsize=20, weight='bold')

# --- Chart 1: Net Profit Growth (Bar Chart) ---
sns.barplot(ax=axes[0, 0], data=df, x='Year', y='Net_Profit', hue='Bank', palette='viridis')
axes[0, 0].set_title('Net Profit Growth (Million $)', fontsize=14)
axes[0, 0].set_ylabel('Net Profit ($M)')
axes[0, 0].grid(axis='y', alpha=0.3)

# --- Chart 2: Customer Deposits Growth (Line Chart) ---
sns.lineplot(ax=axes[0, 1], data=df, x='Year', y='Deposits', hue='Bank', marker='o', linewidth=3, palette='viridis')
axes[0, 1].set_title('Customer Deposits Growth (Billion $)', fontsize=14)
axes[0, 1].set_ylabel('Deposits ($B)')
axes[0, 1].set_xticks([2022, 2023, 2024]) # Ensure only whole years are shown
axes[0, 1].grid(True, alpha=0.3)

# --- Chart 3: Risk Analysis / NPL Ratio (Point Plot) ---
# NPL = Non-Performing Loans. Lower is better.
sns.pointplot(ax=axes[1, 0], data=df, x='Year', y='NPL_Ratio', hue='Bank', palette='magma', linestyle='--')
axes[1, 0].set_title('Risk Analysis: Non-Performing Loans (NPL %)', fontsize=14, color='darkred')
axes[1, 0].set_ylabel('NPL Ratio (%)')
axes[1, 0].grid(True, alpha=0.3)

# --- Chart 4: Executive Summary (Text) ---
axes[1, 1].axis('off')  # Turn off axis for the text box
summary_text = (
    " EXECUTIVE SUMMARY:\n\n"
    "1. Market Dominance: Al Rajhi Bank leads in both Profitability\n"
    "   and Deposit base, indicating massive scale in the KSA market.\n\n"
    "2. Growth Trajectory: KFH (Kuwait) shows the fastest relative\n"
    "   growth rate in profits (+110% over 2 years).\n\n"
    "3. Risk Profile: Al Rajhi maintains the lowest risk (0.4% NPL),\n"
    "   indicating a very healthy loan book. DIB carries higher risk\n"
    "   (1.8%) but is showing a positive improving trend.\n\n"
    " Analyst Recommendation: 'Buy' Al Rajhi for stability,\n"
    "   and KFH for aggressive growth potential."
)
axes[1, 1].text(0.1, 0.5, summary_text, fontsize=13, family='monospace', linespacing=1.6,
                bbox=dict(facecolor='#f0f0f0', alpha=0.5, boxstyle='round,pad=1'))

# 3. Save and Show
plt.savefig('Islamic_Banking_Dashboard.png', dpi=300)
print("Dashboard generated and saved as 'Islamic_Banking_Dashboard.png'")
plt.show()
```
